## Checking the work environment
We check the GitHub repository, the Colab notebook, and access the dataset hosted on the GitHub repo.

```python
import pandas as pd

# load dataset from GitHub
url = "https://raw.githubusercontent.com/Jokiba/ml-predicted-category-product-final-project/main/data/products.csv"
df = pd.read_csv(url)

print("Number of rows:", len(df))
print("First five rows:")
print(df.head())
```

## Checking the dimensions and basic information of the dataset

Here I will check the number of rows and columns of the dataset.<br>
I'll also show its column names and column types.

```python
# Dimension of dataset (number of rows and columns)
print("Dataset shape (rows, columns):", df.shape)

# Standard information about dataset
print("\nDataset info:")
df.info()
```

## Mising values per columns

We check how many non-null (missing) values the columns have.

```python
# Show number of missing values per column
print("Missing values per column:")
df.isna().sum()
```

## Visualize missing data
In this step, using matplotlib and seaborn, we create a heatmap.<br>Heatmap shows us where the missing values are ​​by column of the dataframe.

```python
# Visualize missing data with seaborn heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(13, 5))
sns.heatmap(df.isna(), cbar= False, cmap= "YlOrRd")
plt.title("Missing Values Heatmap")
plt.show()
```

## Products category in numbers
Here we will see what the product category distribution is like.<br>All product categories in one column and their number in another column.

```python
# Count occurrences of each category label
category_counts = df[" Category Label"].value_counts()

# Show number of values column - "Category Label"
print("Product category distribution (counts):")
print(category_counts)
```

## Products category visualization
Visualization of distribution using a bar chart.<br>In this way, we will see if there is a balance between classes or if there is an imbalance in certain classes (categories) of products.

```python
# Plot products category distribution

plt.figure(figsize=(12, 4))
plt.bar(category_counts.index, category_counts.values, color=["skyblue", "salmon"])
plt.xticks(rotation= 45)
plt.title("Products Category Distribution")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.show()
```

## Products category analysis

By analyzing the bar chart, we concluded:


*   The category "Fridge Freezers" has the most products

*   Most categories have good balance
*   Three categories each have under 3000 products
*   Standardization of category name values ​​will be required

*   As well as the standardization of column names
